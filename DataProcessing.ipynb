```python
def custom_print(aut):
    text = ""
    bdict = aut.get_dict()
    text = text + "Acceptance:" + str(aut.get_acceptance()) + ' \n '
    text = text + "Number of sets:" + str(aut.num_sets()) + ' \n '
    text = text + "Number of states:" + str(aut.num_states()) + ' \n '
    text = text + "initial states" + str(aut.get_init_state_number()) + ' \n '
    text = text + "Atomic propositions:"
    for ap in aut.ap():
        text = text + ' ' + str(ap) + ' (=' + str(bdict.varnum(ap)) + ')'
    text = text + ' \n '
    # Templated methods are not available in Python, so we cannot
    # retrieve/attach arbitrary objects from/to the automaton.  However the
    # Python bindings have get_name() and set_name() to access the
    # "automaton-name" property.
    name = aut.get_name()
    if name:
        text = text + "Name: " + name + ' \n '
    return text
    #print("Deterministic:", aut.prop_universal() and aut.is_existential())
    #print("Unambiguous:", aut.prop_unambiguous())
    #print("State-Based Acc:", aut.prop_state_acc())
    #print("Terminal:", aut.prop_terminal())
    #print("Weak:", aut.prop_weak())
    #print("Inherently Weak:", aut.prop_inherently_weak())
    #print("Stutter Invariant:", aut.prop_stutter_invariant())

    #for s in range(0, aut.num_states()):
     #   print(f"State {s}:")
      #  for t in aut.out(s):
       #     print(f"  edge({t.src} -> {t.dst})")
            # bdd_print_formula() is designed to print on a std::ostream, and
            # is inconvenient to use in Python.  Instead we use
            # bdd_format_formula() as this simply returns a string.
        #    print("    label =", spot.bdd_format_formula(bdict, t.cond))
         #   print("    acc sets =", t.acc)
```

```python
def pastConcat(x):
    ran = (x*2) + 1
    #a = spot.automaton('output/arbiter/past/2/past2_0.hoa')
    a = spot.automaton('output/arbiter/past/' + str(x) + '/past' + str(x) + '_0.hoa')
    for y in range(1,ran):
        b = spot.automaton('output/arbiter/past/' + str(x) +'/past' + str(x) + '_' + str(y) +'.hoa')
        a = spot.product(a,b).postprocess('GeneralizedBuchi', 'low', 'complete')
        print(y)
    return a
```

```python
def futureConcat(x):
    ran = (x*2) + 1
    #a = spot.automaton('output/arbiter/past/2/past2_0.hoa')
    a = spot.automaton('output/arbiter/future/' + str(x) + '/future' + str(x) + '_0.hoa')
    for y in range(1,ran):
        b = spot.automaton('output/arbiter/future/' + str(x) +'/future' + str(x) + '_' + str(y) +'.hoa')
        a = spot.product(a,b).postprocess('GeneralizedBuchi', 'low', 'complete')
        print(y)
    return a
```

```python
for i in range(2,6):
    text = custom_print(pastConcat(i))
    with open('Processed/arbiter/past/arbiter' + str(i) + 'processed.txt',"w") as f:
        f.write(text)
```

```python
for i in range(2,6):
    text = custom_print(futureConcat(i))
    with open('Processed/arbiter/future/arbiter' + str(i) + 'processed.txt',"w") as f:
        f.write(text)
```
