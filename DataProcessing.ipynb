```python
def pastConcat(x):
    ran = (x*2) + 1
    a = spot.automaton('output/arbiter/past/' + str(x) +'/past' + str(x) + '_0.hoa')
    for y in range(1,ran):
        b = spot.automaton('output/arbiter/past/' + str(x) +'/past' + str(x) + '_' + str(y) +'.hoa')
        a = spot.product(a,b).postprocess('GeneralizedBuchi', 'low', 'complete')
    return a
```
